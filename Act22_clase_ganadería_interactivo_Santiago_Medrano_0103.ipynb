# Clase Ganadero

```python
# Definición de la clase
class Ganadero:
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    def arrear(self):
        print(f"-> {self.nombre} está arreando al ganado hacia el corral.")

    def alimentar(self):
        print(f"-> {self.nombre} está repartiendo el forraje para los animales.")

    def cuidar_ganado(self):
        print(f"-> {self.nombre} está revisando que el ganado esté sano y seguro.")

# --- Lógica Interactiva ---

lista_ganaderos = []

print("=== SISTEMA DE REGISTRO DE GANADEROS ===")
# Preguntamos cuántos registros se desean hacer
try:
    cantidad = int(input("¿Cuántos ganaderos desea registrar? "))
except ValueError:
    print("Error: Por favor ingrese un número entero.")
    cantidad = 0

for i in range(cantidad):
    print(f"\nDatos del Ganadero {i+1}:")
    nombre = input("Nombre: ")
    telefono = input("Teléfono: ")

    # Validación simple para la edad
    try:
        edad = int(input("Edad: "))
    except ValueError:
        print("Edad no válida, se asignará 0 por defecto.")
        edad = 0

    # Creamos el objeto y lo guardamos en una lista
    nuevo_ganadero = Ganadero(nombre, telefono, edad)
    lista_ganaderos.append(nuevo_ganadero)

# --- Uso de los objetos registrados ---
print("\n" + "="*40)
print("EJECUTANDO TAREAS DIARIAS")
print("="*40)

for g in lista_ganaderos:
    print(f"\nFicha de: {g.nombre.upper()}")
    print(f"Contacto: {g.telefono} | Edad: {g.edad} años")

    # Llamada a sus métodos
    g.arrear()
    g.cuidar_ganado()
    g.alimentar()

print("\n" + "-"*30)
print("Programa terminado con éxito.")
print("Creador de software por: Santiago Medrano NL = 37")
```
